```python
import pandas as pd
from collections import defaultdict
# print asn_countries

# Convert to DataFrame, sort in alphabetical order
asn_countries_df = pd.DataFrame(list(asn_countries.items()), columns=['asn', 'country'])

asn_countries_df.head()

asn_count_per_country_all = defaultdict(int)
asn_count_per_country_higher = defaultdict(int)
asn_count_per_country_lower = defaultdict(int)

for index, row in asn_countries_df.iterrows():
    if row['asn'] in asn_normalized.keys():
        asn_count_per_country_all[row['country']] += 1
        if asn_normalized[row['asn']] > 0.19:
            asn_count_per_country_higher[row['country']] += 1
        else:
            asn_count_per_country_lower[row['country']] += 1

asn_count_per_country_all_df = pd.DataFrame(list(asn_count_per_country_all.items()), columns=['country', 'total_asns'])
asn_count_per_country_higher_df = pd.DataFrame(list(asn_count_per_country_higher.items()), columns=['country', '>5%_asns'])
asn_count_per_country_lower_df = pd.DataFrame(list(asn_count_per_country_lower.items()), columns=['country', '<5%_asns'])

# print asn_count_per_country_all_df
# print asn_count_per_country_higher_df
# print asn_count_per_country_lower_df

asn_count_per_country_df = asn_count_per_country_all_df.set_index('country').join(asn_count_per_country_higher_df.set_index('country')).join(asn_count_per_country_lower_df.set_index('country')).fillna(0)

print(asn_count_per_country_df[asn_count_per_country_df.index == 'NL'].sum())
print(asn_count_per_country_df[asn_count_per_country_df.index != 'NL'].sum())
print(asn_count_per_country_df.sum())
```
